```python
%matplotlib inline
import numpy as np
import torch
from mpl_toolkits import mplot3d
from d2l import torch as d2l
```

```python
def f(x):
    return x * torch.cos(np.pi * x)
def g(x):
    return f(x) + 0.2 * torch.cos(5 * np.pi * x)
```

```python
x = torch.arange(0.5, 1.5, 0.01)
d2l.plot(x, [f(x), g(x)], 'x', 'risk')
d2l.plt.gca().annotate('min of \nempirical risk', xy=(1.0, -1.2), xytext=(0.5, -1.1), arrowprops=dict(arrowstyle='->'))
```
